```python
#18-Elabore uma função chamada calcula_desconto que receba dois parâmetros:
#o valor de um produto e a porcentagem de desconto. A função deve retornar o
#valor final do produto com o desconto aplicado. Teste a função e
#imprima o resultado.
def calcula_desconto(valor_produto, porcentagem_desconto):
    desconto = valor_produto * (porcentagem_desconto / 100)
    valor_final = valor_produto - desconto
    return valor_final

# Teste 1: Produto de R$ 100 com 10% de desconto
preco1 = 100.0
desconto1 = 10
resultado1 = calcula_desconto(preco1, desconto1)
print(f"Produto: R$ {preco1:.2f} | Desconto: {desconto1}% | Preço Final: R$ {resultado1:.2f}")

# Teste 2: Produto de R$ 250.50 com 20% de desconto
preco2 = 250.50
desconto2 = 20
resultado2 = calcula_desconto(preco2, desconto2)
print(f"Produto: R$ {preco2:.2f} | Desconto: {desconto2}% | Preço Final: R$ {resultado2:.2f}")

# Teste 3: Produto de R$ 50 com 0% de desconto
print(f"Produto: R$ 50.00  | Desconto: 0%  | Preço Final: R$ {calcula_desconto(50, 0):.2f}")
```

```python
#19-Elabore uma função chamada classifica_nota que receba uma nota de 0 a 10.
#A função deve retornar "Aprovado" se a nota for maior ou igual a 7,
#"Recuperação" se estiver entre 5 e 6.9, e "Reprovado" se for menor que 5.
#Chame a função passando uma nota e imprima o retorno.

def classifica_nota (nota):
  if nota >= 7:
    return "Aprovado"
  elif nota >= 5 and nota <= 7:
    return "Recuperação"
  else:
    return "Reprovado"

nota1 = 8.5
nota2 = 6.0
nota3 = 4.0
print(f"O aluno com a {nota1} foi: {classifica_nota(nota1)}")
print(f"O aluno com a {nota2} foi: {classifica_nota(nota2)}")
print(f"O aluno com a {nota3} foi: {classifica_nota(nota3)}")
```

```python
#20-Elabore um código que receba uma lista de números inteiros mistos
 #(positivos, negativos e zeros). Utilize um laço for com estruturas
 #condicionais if/elif/else para contabilizar quantos números são positivos,
 #quantos são negativos e quantos são exatamente zero. Ao final,
 #imprima esse resumo.
numeros = [10, -5, 0, 3, -1, 0, 7, -8, 0, 2]

positivos = 0
negativos = 0
zeros = 0

for num in numeros:

    if num > 0:
        positivos += 1
    elif num < 0:
        negativos += 1
    else:
        zeros += 1

# Exibindo o resumo
print("--- Resumo dos Números ---")
print(f"Quantidade de positivos: {positivos}")
print(f"Quantidade de negativos: {negativos}")
print(f"Quantidade de zeros: {zeros}")

```
